```python
import numpy as np
import scipy as sp
import scipy.signal as signal
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

# Questão 5.6-7

```python
b0 = 0.463
b1 = -0.749
b2 = 0.463
a0 = 0.382
a1 = -0.587
a2 = 0.344

w = np.linspace(-3 * np.pi, 3 * np.pi, 10000)
z = np.exp(1j * w)

h = (b0*(z**2) + b1*z + b2) / (a0*(z**2) + a1*z + a2)

plt.plot(w, abs(h))
plt.title('Resposta em Frequência de H(z) - Magnitude')
plt.xlabel('Ω')
plt.ylabel('Magnitude')
plt.show()

plt.plot(w, np.angle(h, deg=True))
plt.title('Resposta em Frequência de H(z) - Fase')
plt.xlabel('ω')
plt.ylabel('Fase (Graus)')
```

# Questão 5.7-2.a

```python
num_s = [7, 20]
den_s = [2, 14, 20]

T = 0.1
num_z, den_z = signal.bilinear(num_s, den_s, fs=1/T)

w, h = signal.freqz(num_z, den_z, worN=10000)

plt.plot(w, np.abs(h))
plt.title('Resposta em Frequência do Filtro Digital')
plt.xlabel('Frequência (rad/s)')
plt.ylabel('Ganho (dB)')
plt.show()
```
