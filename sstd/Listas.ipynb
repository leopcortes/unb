# Funções Básicas

```python
def decimacao(sinal, fator):
    decimado = []
    
    for i in range(len(sinal)):
        if i % fator == 0:
            decimado.append(sinal[i])
    
    return np.array(decimado)
  
def interpolacao(sinal, fator):
    interpolado = []
    
    for i in range(len(sinal)*fator):
        if i % fator == 0:
            interpolado.append(sinal[i//fator])
        else:
            interpolado.append(interpolado[i-1])
    
    return np.array(interpolado)

def delta(n):
    if n == 0:
        return 1
    else:
        return 0

def u(n):
    if n >= 0:
        return 1
    else:
        return 0

def energia(sinal):
    energ = 0.0
    for n in range(len(sinal)):
        energ += abs(sinal[n])**2
    
    return energ
    
def potencia(sinal):
    pot = sum(abs(x)**2 for x in sinal) / (len(sinal))
    
    return pot
```

```python
def fibonacci(n):
  if n <= 1:
    return n
  else:
    return fibonacci(n-1) + fibonacci(n-2)
  
fib = ((1/math.sqrt(5))*((1 + math.sqrt(5))/2)**49) - ((1/math.sqrt(5))*((1 - math.sqrt(5))/2)**49)
print(fib)
```

# Convolução

```python
def sistema(n):
  if n < 1:
    return 0
  else:
    return (-2)**n

def entrada(n):
  if n < -1:
    return 0
  else:
    return math.e**(-n)

def convolucao(h, x):
  y = []
  for i in range(len(h) + len(x)):
    elemento = 0
    for j in range(len(h) + len(x)):
      if i - j >= 0:
        if (j > (len(h)-1)) and ((i-j) > (len(x)-1)):
          elemento += 0 * 0
        elif j > (len(h)-1):
          elemento += 0 * x[i-j]
        elif (i-j) > (len(x)-1):
          elemento += h[j] * 0
        else:
          elemento += h[j] * x[i-j]
    y.append(elemento) 
  return y

h, g = [], []

for n in range(20):
  h.append(sistema(n))
  g.append(entrada(n))
  
sinal_convoluido = convolucao(h, g)

for i in range(20):
    sinal_convoluido.append(sinal_convoluido[-1])

plt.plot(range(20), h)
plt.title("Gráfico de h[n]")
plt.xlabel("n")
plt.ylabel("sinal")
plt.show()

plt.plot(range(20), g)
plt.title("Gráfico de g[n]")
plt.xlabel("n")
plt.ylabel("sinal")
plt.show()

plt.plot(sinal_convoluido)
plt.title("Convolução de h[n] e x[n] obtida por função direta")
plt.xlabel("n")
plt.ylabel("sinal")
plt.show()
```
