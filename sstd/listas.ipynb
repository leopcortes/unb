```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
from PIL import Image
import math
import cmath
import warnings
warnings.filterwarnings('ignore')
```

```python
def decimacao(sinal, fator):
    decimado = []
    
    for i in range(len(sinal)):
        if i % fator == 0:
            decimado.append(sinal[i])
    
    return np.array(decimado)
  
def interpolacao(sinal, fator):
    interpolado = []
    
    for i in range(len(sinal)*fator):
        if i % fator == 0:
            interpolado.append(sinal[i//fator])
        else:
            interpolado.append(interpolado[i-1])
    
    return np.array(interpolado)

def delta(n):
    if n == 0:
        return 1
    else:
        return 0

def u(n):
    if n >= 0:
        return 1
    else:
        return 0

def energia(sinal):
    energ = 0.0
    for n in range(len(sinal)):
        energ += abs(sinal[n])**2
    
    return energ
    
def potencia(sinal):
    pot = sum(abs(x)**2 for x in sinal) / (len(sinal))
    
    return pot
```

```python
def fibonacci(n):
  if n <= 1:
    return n
  else:
    return fibonacci(n-1) + fibonacci(n-2)
  
fib = ((1/math.sqrt(5))*((1 + math.sqrt(5))/2)**49) - ((1/math.sqrt(5))*((1 - math.sqrt(5))/2)**49)
print(fib)
```
