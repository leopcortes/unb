```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
from PIL import Image
import math
import cmath
import warnings
warnings.filterwarnings('ignore')
```

```python
def decimacao(sinal, fator):
    decimado = []
    
    for i in range(len(sinal)):
        if i % fator == 0:
            decimado.append(sinal[i])
    
    return np.array(decimado)
  
def interpolacao(sinal, fator):
    interpolado = []
    
    for i in range(len(sinal)*fator):
        if i % fator == 0:
            interpolado.append(sinal[i//fator])
        else:
            interpolado.append(interpolado[i-1])
    
    return np.array(interpolado)

def delta(n):
    if n == 0:
        return 1
    else:
        return 0

def u(n):
    if n >= 0:
        return 1
    else:
        return 0

def energia(sinal):
    energ = 0.0
    for n in range(len(sinal)):
        energ += abs(sinal[n])**2
    
    return energ
    
def potencia(sinal):
    pot = sum(abs(x)**2 for x in sinal) / (len(sinal))
    
    return pot
```

```python
h = []

for i in range(-10, 10):
  h.append(4*delta(i+2) + 2*delta(i+1) + 2*delta(i-1) + 4*delta(i-2))
  
plt.stem(range(-10, 10), h)
plt.title("Gráfico de h[n]")
plt.xlabel("n")
plt.ylabel("sinal")
plt.show()
```
